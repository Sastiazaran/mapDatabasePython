```python
houses = pd.read_csv('houses.csv')
stores = pd.read_csv('plazas.csv')
hospitals = pd.read_csv('hospitales.csv')
universities = pd.read_csv('Universidades.csv')
sports = pd.read_csv('sports.csv')
tourisms = pd.read_csv('turismo.csv')
eco_tourisms = pd.read_csv('turismo_ecologico.csv')

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

houses.head(2400)
```

```python
my_map = folium.Map(
    location = [20.72356, -103.38479]
    
)
my_map

for _, house in houses.iterrows():
    folium.Marker(
        location = [house['latitud'], house['longitud']],
        tiles = "Stamen Terrain",
        popup=house['price'],
        tooltip=house['direction'],
        icon = folium.Icon( icon = 'home', prefix='fa')
        
    ).add_to(my_map)
    
for _, store in stores.iterrows():  
    folium.Marker(
        location = [store['latitud'], store['longitud']],
        tooltip= store['direction'],
        icon = folium.Icon (icon = 'shopping-bag', prefix='fa', color = 'purple')
        
    
    ).add_to(my_map)
    
for _, hospital in hospitals.iterrows():
    folium.Marker(
        location = [hospital['latitud'], hospital['longitud']],
        tooltip = hospital['direction'],
        icon = folium.Icon (icon = 'h-square', prefix = 'fa', color = 'lightblue')
    
    ).add_to(my_map)
    
for _, university in universities.iterrows():
    folium.Marker(
        location = [university['latitud'], university['longitud']],
        tooltip = university['direction'],
        icon = folium.Icon (icon = 'university', prefix = 'fa', color = 'darkpurple')
    ).add_to(my_map)
    
for _, sport in sports.iterrows():
    folium.Marker(
        location = [sport['latitud'], sport['longitud']],
        tooltip = sport['direction'],
        icon = folium.Icon(icon = 'bitbucket', prefix = 'fa', color = 'beige')
        
    ).add_to(my_map)
    
for _, tourism in tourisms.iterrows():
    folium.Marker(
        location = [tourism['latitud'], tourism['longitud']],
        tooltip = tourism['direction'],
        icon = folium.Icon(icon = 'map-pin', prefix = 'fa', color = 'pink')
        
    ).add_to(my_map)
    
for _, tourism in tourisms.iterrows():
    folium.Marker(
        location = [tourism['latitud'], tourism['longitud']],
        tooltip = tourism['direction'],
        icon = folium.Icon(icon = 'map-pin', prefix = 'fa', color = 'pink')
        
    ).add_to(my_map)
    
for _, eco_tourism in eco_tourisms.iterrows():
    folium.Marker(
        location = [eco_tourism['latitud'], eco_tourism['longitud']],
        tooltip = eco_tourism['direction'],
        icon = folium.Icon(icon = 'tree', prefix = 'fa', color = 'lightgreen')
        
    ).add_to(my_map)


my_map
```

```python
def select_marker_color(row):
    if row['price'] <= 2000000:
        return 'red'
    elif row['price'] >= 2000001 and row['price'] <= 7000000:
        return 'blue'
    else:
        return 'green'
```

```python
houses['color'] = houses.apply(select_marker_color, axis=1)
houses.head(3)
```

```python
my_map = folium.Map(
    location = [20.72356, -103.38479]
    
)
my_map

for _, house in houses.iterrows():
    folium.Marker(
        location = [house['latitud'], house['longitud']],
        popup=house['price'],
        tooltip=house['direction'],
        icon=folium.Icon(color=house['color'], icon = 'home')
        
    ).add_to(my_map)
    
for _, store in stores.iterrows():  
    folium.Marker(
        location = [store['latitud'], store['longitud']],
        tooltip= store['direction'],
        icon = folium.Icon (icon = 'shopping-bag', prefix='fa', color = 'purple')
    ).add_to(my_map)
        
for _, hospital in hospitals.iterrows():
    folium.Marker(
        location = [hospital['latitud'], hospital['longitud']],
        tooltip = hospital['direction'],
        icon = folium.Icon (icon = 'h-square', prefix = 'fa', color = 'lightblue')
    
    
    ).add_to(my_map)
    
for _, university in universities.iterrows():
    folium.Marker(
        location = [university['latitud'], university['longitud']],
        tooltip = university['direction'],
        icon = folium.Icon (icon = 'university', prefix = 'fa', color = 'darkpurple')
    ).add_to(my_map)
    
for _, sport in sports.iterrows():
    folium.Marker(
        location = [sport['latitud'], sport['longitud']],
        tooltip = sport['direction'],
        icon = folium.Icon(icon = 'bitbucket', prefix = 'fa', color = 'beige')
        
    ).add_to(my_map)
    
for _, tourism in tourisms.iterrows():
    folium.Marker(
        location = [tourism['latitud'], tourism['longitud']],
        tooltip = tourism['direction'],
        icon = folium.Icon(icon = 'map-pin', prefix = 'fa', color = 'pink')
        
    ).add_to(my_map)


my_map
```

```python
def select_marker_color(row):
    if row['size'] <= 150:
        return 'red'
    elif row['size'] >= 151 and row['size'] <= 300:
        return 'blue'
    else:
        return 'orange'
```
